## Downloading data

```python
imgs = pd.read_csv('../FEC_dataset/faceexp-comparison-data-test-public.csv',sep='\s+',header=None)
imgs.columns=['Image']
```

```python
imgs.head()
```

```python
imgs = imgs.Image.str.split(',',expand=True,)
imgs = imgs[[0]]
```

```python
def url_to_image(url):
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image
```

```python
prev = imgs[0][0]
i = 1
for url in tqdm(imgs[0]):
    if url != prev:
        try:
            img = url_to_image(url)
            i += 1
            print(i)
            prev = url
            cv2.imwrite('./data/test/'+'image_real_'+str(i)+'.jpg', img)
        except:
            prev = url
```
